## Linear Dimensionality Reduction : PCA

### Just skills data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
pca_data = data[skills]
min_max_scaler = MinMaxScaler()
pca_data = min_max_scaler.fit_transform(pca_data)
```

```python
components_to_try = list(range(2, 100))

for component in components_to_try:
  pca = PCA(n_components=component)
  pca.fit(pca_data)
  print(pca.explained_variance_ratio_.sum())
```

### PCA of the skills + StudentAgeAtAssesment + IsMale + FirstAssessment_byStudent

```python
non_skill_useful_features = ['FirstAssessment_byStudent', 'StudentAgeAtAssesment', 'IsMale']
useful_features = list(skills) + non_skill_useful_features
```

 Note that 1381 rows don't have Age and 374 rows don't have Gender.
 We drop them to proceed

```python
pca_data = data[useful_features].dropna()
pca_data = min_max_scaler.fit_transform(pca_data)
components_to_try = list(range(2, 100))

for component in components_to_try:
  pca = PCA(n_components=component)
  pca.fit(pca_data)
  print(pca.explained_variance_ratio_.sum())
```

### PCA by skill area

```python
skill_areas = list(string.ascii_uppercase)
skill_areas.remove('O')
```

```python
for skill_area in skill_areas:
  print("#########  SKILL AREA : " + skill_area)
  components_to_try = list(range(2, 6))
  skills_by_area = [x for x in skills if skill_area in x]

  pca_data_by_area = data[skills_by_area]
  pca_data_by_area = min_max_scaler.fit_transform(pca_data_by_area)

  for component in components_to_try:
    pca = PCA(n_components=component)
    pca.fit(pca_data_by_area)
    print(pca.explained_variance_ratio_.sum())
```

## Non-Linear Dimensionality Reduction : t-SNE

### Just skills data

```python
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
```

```python
tsne_data = data[skills]
tsne_data = min_max_scaler.fit_transform(tsne_data)
```

```python
tsne_data_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3).fit_transform(tsne_data)
plt.figure(figsize=(6, 5))
plt.scatter(tsne_data_embedded[:,0], tsne_data_embedded[:,1], s = 1)
plt.show()
```

```python
perplexities = [5, 30, 60, 100]
iterations = [1000, 2500, 5000]
(fig, subplots) = plt.subplots(4, 3, figsize=(24, 24))

for i, perplexity in enumerate(perplexities):
  for j, iter in enumerate(iterations):
    ax = subplots[i][j]
    ax.set_title("Perplexity=%d" % perplexity + "\nIteration=%d" % iter)
    tsne_data_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=perplexity, n_iter=iter).fit_transform(tsne_data)
    ax.scatter(tsne_data_embedded[:,0], tsne_data_embedded[:,1], s=1)
```

*It makes more sense to use t-SNE to visualise the clusters after the after the Clustering Step.*
